```python
import sympy as sym
import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np
mpl.rcParams['figure.figsize'] = (10,10)
```

```python
x, y = sym.symbols('x y')
```

```python
f = sym.sin(x)*sym.exp(-x/4)
f
```

```python
g = (x-1)**2
sym.expand(g)
```

```python
dfdx = sym.diff(f,x)
dfdx
```

```python
sym.integrate(f,(x,0,1))
```

```python
f.subs(x,1)
```

```python
f.subs(x,1).evalf()
```

```python
u = np.linspace(0,10,100)
f_lam = sym.lambdify(x,f)
dfdx_lam = sym.lambdify(x,dfdx)
plt.plot(u, f_lam(u), '-r', u, dfdx_lam(u), '-g')
plt.grid(True)
plt.legend(['f(x)', 'df(x)/dx'])
```
